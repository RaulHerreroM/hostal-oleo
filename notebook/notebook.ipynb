```python
import requests
import json
import pandas as pd

# Definir los headers y los datos de la solicitud
headers = {'x-api-key': 'wb_06064db8-fce1-11ea-90d0-001a4a4ef9b1'}
created = {'from': '10/10/2021', 'to': '10/12/2024'}
arrival = {'from': '08/11/2021', 'to': '10/12/2024'}
all_reservations = []
limit = 64
offset = 0
has_more_reservations = True

while has_more_reservations:
    # Crear el cuerpo de la solicitud con el offset actualizado
    pager = {'limit': limit, 'offset': offset}
    data = {'created': created, 'arrival': arrival, 'pager': pager}
    data = {'filters': json.dumps(data)}
    
    # Hacer la solicitud POST
    response = requests.post('https://kapi.wubook.net/kp/reservations/fetch_reservations', headers=headers, data=data)
    
    # Comprobar si la solicitud fue exitosa
    if response.status_code == 200:
        # Obtener los datos de la respuesta
        response_json = response.json()
        
        # Extraer las reservas de la respuesta JSON
        reservations = response_json.get('data', {}).get('reservations', [])
        
        # Si no hay ms reservas, terminamos el bucle
        if not reservations:
            has_more_reservations = False
        else:
            # Aadir las reservas obtenidas a la lista total
            all_reservations.extend(reservations)
            
            # Incrementar el offset para la prxima solicitud
            offset += limit
    else:
        print(f"Error en la solicitud: {response.status_code}")
        has_more_reservations = False

# Crear una lista para almacenar los datos procesados
reservations_data = []

for reservation in all_reservations:
    reservation_id = reservation.get('id')
    status = reservation.get('status')
    origin = reservation.get('origin', {}).get('channel')
    created = reservation.get('created')
    total_price = reservation.get('price', {}).get('total')
    
    # Contar el nmero de habitaciones reservadas
    n_rooms = len(reservation.get('rooms', []))
    
    # Extraer las fechas dfrom y dto de la primera habitacin (si existe)
    if n_rooms > 0:
        dfrom = reservation.get('rooms')[0].get('dfrom')
        dto = reservation.get('rooms')[0].get('dto')
    else:
        dfrom = None
        dto = None
        
    reservations_data.append({
        'id': reservation_id,
        'status': status,
        'origin': origin,
        'created': created,
        'n_rooms': n_rooms,
        'dfrom': dfrom,
        'dto': dto,
        'total_price': total_price
    })

# Crear un DataFrame a partir de los datos de las reservas
df = pd.DataFrame(reservations_data)

# Mostrar el DataFrame
print(df)

# Crear un DataFrame a partir de los datos de las reservas
df = pd.DataFrame(reservations_data)

# Mostrar el DataFrame
print(df)
```

```python
import pandas as pd
df=df[df['status']=='Confirmed']

# Convertir las columnas de fechas en formato de fecha
df['created'] = pd.to_datetime(df['created'], format='%d/%m/%Y')
df['dfrom'] = pd.to_datetime(df['dfrom'], format='%d/%m/%Y')
df['dto'] = pd.to_datetime(df['dto'], format='%d/%m/%Y')

# Crear un rango de das desde la fecha ms temprana de 'dfrom' hasta la ms tarda de 'dto'
min_date = df['dfrom'].min()
max_date = df['dto'].max()

# Generar un DataFrame que tiene una fila por cada da en ese rango
df_days = pd.DataFrame({'date': pd.date_range(start=min_date, end=max_date)})
```

```python
# Funcin para calcular cuntas habitaciones estn ocupadas en cada da
def get_occupied_rooms(day, df):
    occupied_rooms = df[(df['status'] == 'Confirmed') & (df['dfrom'] <= day) & (df['dto'] > day)]
    return occupied_rooms['n_rooms'].sum()

# Aplicar la funcin para cada da en el DataFrame df_days
df_days['occupied_rooms'] = df_days['date'].apply(lambda day: get_occupied_rooms(day, df))
```

```python
def get_avg_price_per_room_and_day(day, df):
    # Filtrar las reservas que estn activas en el da dado
    occupied_rooms = df[(df['status'] == 'Confirmed') & (df['dfrom'] <= day) & (df['dto'] >= day)]
    
    total_price = 0
    total_rooms = 0
    
    for _, row in occupied_rooms.iterrows():
        num_days = (row['dto'] - row['dfrom']).days  # Duracin de la estancia en das
        total_price += row['total_price'] / num_days  # Precio por da
        total_rooms += row['n_rooms']  # Contar las habitaciones
    
    if total_rooms == 0:
        return 0  # Evitar divisin por 0
    
    return total_price / total_rooms  # Precio medio por habitacin y por da

# Aplicar la funcin para calcular el precio medio por habitacin y por da para cada fecha
df_days['avg_price_per_room'] = df_days['date'].apply(lambda day: get_avg_price_per_room_and_day(day, df))
```

```python
from datetime import timedelta

# Definir periodos de anticipacin
anticipation_periods = [1, 2, 3, 5, 7, 14, 30, 90]

# Crear una columna para cada periodo de anticipacin
for days in anticipation_periods:
    df_days[f'rooms_reserved_{days}_days_before'] = df_days['date'].apply(
        lambda day: df[
            (df['status'] == 'Confirmed') &
            (df['dfrom'] == day) &  # Fecha de entrada del cliente es el da en cuestin
            (df['created'] == day - timedelta(days=days))  # La reserva se hizo 'days' das antes
        ]['n_rooms'].sum()
    )
```

```python
# Importar libreras necesarias
import numpy as np

# Crear una copia de la columna original para preservarla si es necesario
df_days['avg_price_per_room_filled'] = df_days['avg_price_per_room']

# Funcin para reemplazar los valores 0 con el promedio entre el da anterior y posterior
def fill_zero_with_neighbors(df):
    # Iterar sobre los ndices del DataFrame
    for i in range(len(df)):
        if df.loc[i, 'avg_price_per_room_filled'] == 0:
            # Buscar el da anterior que no tenga un valor 0
            prev_val = None
            next_val = None
            
            # Buscar el valor anterior que no sea 0
            for j in range(i-1, -1, -1):
                if df.loc[j, 'avg_price_per_room_filled'] != 0:
                    prev_val = df.loc[j, 'avg_price_per_room_filled']
                    break
            
            # Buscar el valor posterior que no sea 0
            for j in range(i+1, len(df)):
                if df.loc[j, 'avg_price_per_room_filled'] != 0:
                    next_val = df.loc[j, 'avg_price_per_room_filled']
                    break
            
            # Si ambos valores existen (anterior y posterior), calcular el promedio
            if prev_val is not None and next_val is not None:
                df.loc[i, 'avg_price_per_room_filled'] = (prev_val + next_val) / 2
            # Si solo existe el valor anterior, usar ese
            elif prev_val is not None:
                df.loc[i, 'avg_price_per_room_filled'] = prev_val
            # Si solo existe el valor posterior, usar ese
            elif next_val is not None:
                df.loc[i, 'avg_price_per_room_filled'] = next_val

# Aplicar la funcin para rellenar los valores 0 en el DataFrame
fill_zero_with_neighbors(df_days)

# Mostrar el resultado final
df_days.head()
```
